# 6. Applying SMOTE to resample Data

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

```python
x_train2,x_test2,y_train2,y_test2 = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Feature Scaling
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train2)
x_test_scaled = scaler.transform(x_test2)
```

```python
# Convert target to 1D
y_train2 =  y_train2.values.ravel()
```

```python
# Apply SMOTE
smote = SMOTE(random_state=42)
x_resampled,y_resampled = smote.fit_resample(x_train_scaled,y_train2)
```

```python
print("Before SMOTE:", pd.Series(y_train2).value_counts())
print("After SMOTE:", pd.Series(y_resampled).value_counts())
```

## 7. Train Model on Resampled Data

#### 7.1 Logisitic Regression

#### A) Tranining of model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_resampled,y_resampled)
```

```python
y_pred_lr2= lr.predict(x_test_scaled)
print("Accuracy:",accuracy_score(y_test2,y_pred_lr2)*100)
```

```python
cr_lr2 = classification_report(y_pred_lr2,y_test2)
print(cr_lr2)
```

#### B) Compare Accuracy Before and After SMOTE

```python
print("Accuracy Before SMOTE is: ",accuracy_score(y_pred_lr1,y_test)*100)
print("Accuracy After SMOTE is: ", accuracy_score(y_test2,y_pred_lr2)*100)
```

#### C) REASON OF Accuracy Decreasing

# 8. Improve Accuracy of RFC Model

#### A) RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV, cross_val_score
```

```python
params = {
    'n_estimators': [200, 300, 500],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False]
}
```

```python
rfc3 = RandomForestClassifier(random_state=42)
rfc3_search = RandomizedSearchCV(rfc3,param_distributions=params,n_iter=15,
              cv=3,scoring='f1_macro',n_jobs=-1,random_state=42)
                                 
```

```python
rfc3_search.fit(x_resampled,y_resampled)
```

```python
print("Best Parameters are: ",rfc3_search.best_params_)
```

```python
best_rfc = rfc3_search.best_estimator_
best_rfc.fit(x_resampled,y_resampled)
y_pred_best = best_rfc.predict(x_test_scaled)
```

```python
print("\n Final Accuracy:",accuracy_score(y_test2,y_pred_best)*100)
print("\n Classification Report:\n",classification_report(y_test2,y_pred_best))
```

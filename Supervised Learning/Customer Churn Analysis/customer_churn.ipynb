### 3.Extract all the male senior citizens whose Payment Method is Electronic check & store the result in ‘senior_male_electronic’

```python
# This isolates a specific customer segment — great for subgroup analysis.
senior_male_electronic = data[(data['gender'] == 'Male') & (data['SeniorCitizen'] == 1) & 
(data['PaymentMethod'] == 'Electronic check')]
senior_male_electronic.sample(5)
```

### 4. Extract all those customers whose tenure is greater than 70 months or their Monthly charges is more than 100$ & store the result in customer_total_tenure

```python
# These could be high-value customers — good for targeted retention or reward analysis.
customer_total_tenure = data[(data['tenure'] > 70) | (data['MonthlyCharges'] > 100)]
customer_total_tenure.sample(3)
```

### 5. Contract: 2 years, Mailed check, Churn: Yes

```python
two_mail_yes = data[(data['Contract'] == 'Two year') &
                    (data['PaymentMethod'] == 'Mailed check') &
                    (data['Churn'] == 'Yes')]
two_mail_yes.sample(3)
```

### 6. Extract 333 random records

```python
customer_333 = data.sample(n = 333, random_state= 42)
customer_333.sample(4)
```

# 4.3 Model Building

## Linear Regression

### Build a simple linear model where dependent variable is ‘MonthlyCharges’ and independent variable is ‘tenure’

#### Split Data

```python
x = data[['tenure']]
y = data[['MonthlyCharges']]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictions = lr.predict(x_test)
```

```python
plt.scatter(predictions,y_test)
```

### Model Evaluation

```python
lr.intercept_
```

```python
lr.coef_
```

```python
MAE = mean_absolute_error(y_test,predictions)
MAE
```

```python
Variance = mean_squared_error(predictions,y_test)
Variance
```

```python
Standard_Deviation = np.sqrt(mean_squared_error(y_test,predictions))
Standard_Deviation
```

```python
r2_score(y_test,predictions)
```

## 4.4 Logistic Regression

### Build a simple logistic regression model where dependent variable is "Churn" & independent variable is "MonthlyCharges"

```python
x = data[['MonthlyCharges']]
y = data[['Churn']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
lgr = LogisticRegression()
lgr.fit(x_train,y_train)
```

```python
pred2 = lgr.predict(x_test)
```

```python
classification_report(pred2,y_test)
```

```python
accuracy_score(pred2,y_test)*100
```

# Build a multiple logistic regression model where dependent variable is ‘Churn’ & independent variables are ‘tenure’ & ‘MonthlyCharges’

#### Data Split

```python
x = data[['MonthlyCharges','tenure']]
y = data[['Churn']]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

#### Building Model 

```python
lgr2 = LogisticRegression()
lgr2.fit(x_train,y_train)
```

```python
predictions = lgr2.predict(x_test)
```

#### Model Evaluation

```python
accuracy_score(predictions,y_test)*100
```

```python
confusion_matrix(predictions,y_test)
```

```python
classification_report(predictions,y_test)
```
